Predict what food item will be served after 10 days for the given state diagram below
State Transistion Matrix **bold text**

```python
import numpy as np

# Define states
states = ["Idli", "Dosa", "Pongal"]

# Transition matrix
transition_matrix = np.array([
    [0.6, 0.3, 0.1],
    [0.2, 0.7, 0.1],
    [0.3, 0.3, 0.4]
])

# Initial state: starting with Idli
initial_state = np.array([1, 0, 0])

# Predict for 10 days
num_days = 10
state_distribution = initial_state.copy()
state_distributions = [state_distribution.copy()]

for _ in range(num_days):
    state_distribution = np.dot(state_distribution, transition_matrix)
    state_distributions.append(state_distribution.copy())

# Print the predictions
print("Predicted food item probabilities each day:")
for day, distribution in enumerate(state_distributions):
    percentages = [f"{prob * 100:.2f}%" for prob in distribution]
    print(f"Day {day}: {dict(zip(states, percentages))}")
```
